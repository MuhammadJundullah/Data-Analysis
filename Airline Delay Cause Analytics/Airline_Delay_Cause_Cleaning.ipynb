# Data Cleaning

```python
df = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/Airline Delay Cause_Dataset/Airline_Delay_Cause.csv")
```

```python
df.info()
```

```python
df.shape
```

### Menyesuaikan tipe Data

```python
 df['date'] = pd.to_datetime(df[['year', 'month']].assign(day=1))
```

```python
df.head()
```

### Memindahkan kolom date

```python
df.insert(0, "date", df.pop("date"))
```

```python
df.head()
```

### Menghapus kolom year dan month

```python
df = df.drop(columns=["year","month"])
df.head()
```

```python
df.dtypes
```

```python
df.tail()
```

### Menyesuaikan format tanggal

```python
df['date'] = df['date'].dt.strftime('%m/%Y')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.head()
```

### Memisahkan kolom airport_name menjadi 2 kolom key dan city

```python
df[['key', 'city']] = df['airport_name'].str.split(":", expand=True)
df.head()
```

### Menghapus kolom airport_name

```python
df.drop(columns=["airport_name"], inplace=True)
```

```python
df.head()
```

### Menyesuaikan posisi kolom key dan city

```python
df.insert(4, "key", df.pop("key"))
df.insert(5, "city", df.pop("city"))
df.head()
```

### Rename kolom city menjadi airport_name

```python
df.rename(columns={"city":"airport_name"}, inplace=True)
df.head()
```

### Rename kolom key menjadi city

```python
df.rename(columns={"key":"city"}, inplace=True)
df.head()
```

### Memisahkan kolom city menjadi kolom city dan state

```python
df[['city', 'state']] = df['city'].str.split(",", expand=True)
df.head()
```

### Menyesuaikan posisi kolom state

```python
df.insert(5, "state", df.pop("state"))
df.head()
```

### Menghapus spasi berlebih dalam data

```python
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df.head()
```

### Memotong data sesuai kebutuhan analisis

```python
df.drop(df.loc[:, 'arr_flights':'arr_diverted'].columns,axis = 1, inplace=True)
df.head()
```

### Eksport data clean

```python
df = df.to_csv("/content/gdrive/MyDrive/Colab Notebooks/Airline Delay Cause_Dataset/Airline_Delay_Cause_clean.csv", index=False)
```
